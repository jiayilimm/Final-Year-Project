```
# news_db = pd.DataFrame(columns=['event','url','date','title','text'])
news_db = pd.read_excel('/content/drive/My Drive/FYP/news_db_22_28.xlsx',index_col=[0])
news_db
```

```
event_list, url_list = [],[]
date_list = []
title_list = []
text_list = []
temp_df = pd.DataFrame(columns=['event','url','date','title','text'])
files = ['/content/drive/MyDrive/FYP/CC-NEWS-20200928171340-00248.warc.gz',
         '/content/drive/MyDrive/FYP/CC-NEWS-20200928182436-00249.warc.gz',
         '/content/drive/MyDrive/FYP/CC-NEWS-20200928193509-00250.warc.gz',
         '/content/drive/MyDrive/FYP/CC-NEWS-20200928210046-00251.warc.gz']
for path in files:
  with open(path, 'rb') as stream:
      for record in ArchiveIterator(stream):
          url = record.rec_headers.get_header('WARC-Target-URI')
          if record.rec_type == 'response':
              if "quake" in url or "magnitude" in url:
                # print(url)
                event_list.append('earthquake')
                url_list.append(url)
              elif "flood" in url:
                # print(url)
                event_list.append('flood')
                url_list.append(url)
              elif "hurricane" in url:
                # print(url)
                event_list.append('hurricane')
                url_list.append(url)
              elif "car-accident" in url or "road-accident" in url or "road-crash" in url or "car-crash" in url or "hit-run" in url or "hit-and-run" in url:
                # print(url)
                event_list.append('road accident')
                url_list.append(url)
              elif "protest" in url or "demonstration" in url:
                # print(url)
                event_list.append('unrest')
                url_list.append(url)
              elif "shoot" in url or "killed" in url or "injured" in url or "attack" in url:
                # print(url)
                event_list.append('assault')
                url_list.append(url)


temp_df['event'] = event_list
temp_df['url'] = url_list
```

```
for url in url_list:
    try:
        article = Article(url, language="en")
        article.download()
        article.parse()
        date = article.publish_date
        #converted to str
        date = '{:%Y-%m-%d}'.format(date)  
        title = article.title
        text = article.text
    except:
        date = np.nan
        title = np.nan
        text = np.nan
    date_list.append(date)
    title_list.append(title)
    text_list.append(text)
# print(date_list)
# print(title_list)
# print(text_list)
temp_df['date'] = date_list
temp_df['title'] = title_list
temp_df['text'] = text_list 
temp_df.dropna(subset=['title'], inplace=True)
```

```
result = pd.concat([news_db,temp_df])
result = result.reset_index(drop=True)
result.to_excel('/content/drive/My Drive/FYP/news_db_22_28.xlsx', sheet_name='news_db')
```
