```
import nltk
import pandas as pd
import re
import spacy
from spacy import displacy 
```

```
nlp = spacy.load("en_core_web_sm")
main_df = pd.read_excel("details.xlsx")
```

```
sentences = main_df['sentence']
```

```
def find_date(text):
  doc = nlp(text)
  dates = [ent.text for ent in doc.ents if ent.label_ == 'DATE']
  if len(dates) != 0:
    return dates
  else:
    pass
```

```
def find_loc(text):
  doc = nlp(text)
  loc = [ent.text for ent in doc.ents if ent.label_ == 'LOC' or 'GPE']
  if len(loc) != 0:
    return loc
  else:
    pass
```

```
def find_time(text):
  doc = nlp(text)
  times = [ent.text for ent in doc.ents if ent.label_ == 'TIME']
  if len(times) != 0:
    return times
  else:
    pass
```

```
for sentence in sentences:
  print('Location Identified: ',find_loc(sentence))
  print('Date Identified: ',find_date(sentence))
  print('Time Identified: ',find_time(sentence))
```

```
def display(text):
  doc = nlp(text)
  options = {'ents':['LOC','GPE','DATE','TIME']}
  displacy.render(doc, style = 'ent', jupyter=True, options=options)

for sentence in sentences:
  display(sentence)
```
